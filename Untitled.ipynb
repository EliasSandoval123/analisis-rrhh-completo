```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar datos
df = pd.read_csv('rrhh_sample.csv')

# Revisar estructura
print(df.shape)
print(df.dtypes)
df.head()
```

```python
# Limpieza y preparación
df['fecha_ingreso'] = pd.to_datetime(df['fecha_ingreso'])
df['fecha_salida'] = pd.to_datetime(df['fecha_salida'], errors='coerce')

# Calcular antigüedad en años
fecha_ref = pd.Timestamp('2024-01-01')
df['antiguedad_anos'] = ((fecha_ref - df['fecha_ingreso']).dt.days / 365).round(1)

# Columna rotado (1 = se fue, 0 = activo)
df['rotado'] = df['fecha_salida'].notna().astype(int)

print(f"Total empleados: {len(df)}")
print(f"Rotados: {df['rotado'].sum()} ({df['rotado'].mean()*100:.1f}%)")
print(f"Activos: {(df['rotado']==0).sum()}")
df[['nombre','antiguedad_anos','rotado']].head(8)
```

```python
# Gráficos de análisis
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Análisis de Rotación de Personal', fontsize=16, fontweight='bold')

# 1. Rotación por departamento
rotacion_dept = df.groupby('departamento')['rotado'].mean() * 100
rotacion_dept.sort_values().plot(kind='barh', ax=axes[0,0], color='#E8593C')
axes[0,0].set_title('Tasa de rotación por departamento (%)')
axes[0,0].set_xlabel('%')

# 2. Satisfacción vs Horas extra
colores = df['rotado'].map({0: '#1D9E75', 1: '#E8593C'})
axes[0,1].scatter(df['horas_extra_mes'], df['satisfaccion'], c=colores, alpha=0.7, s=80)
axes[0,1].set_title('Satisfacción vs Horas extra')
axes[0,1].set_xlabel('Horas extra al mes')
axes[0,1].set_ylabel('Satisfacción (1-5)')
axes[0,1].legend(handles=[
    plt.scatter([],[], c='#E8593C', label='Rotado'),
    plt.scatter([],[], c='#1D9E75', label='Activo')
], loc='upper right')

# 3. Motivos de salida
motivos = df[df['rotado']==1]['motivo_salida'].value_counts()
motivos.plot(kind='bar', ax=axes[1,0], color='#378ADD', rot=15)
axes[1,0].set_title('Motivos de desvinculación')
axes[1,0].set_ylabel('Cantidad')

# 4. Salario promedio por departamento y estado
df_sal = df.groupby(['departamento','rotado'])['salario'].mean().unstack()
df_sal.columns = ['Activo','Rotado']
df_sal.plot(kind='bar', ax=axes[1,1], color=['#1D9E75','#E8593C'], rot=15)
axes[1,1].set_title('Salario promedio: Activos vs Rotados')
axes[1,1].set_ylabel('Salario CLP')

plt.tight_layout()
plt.savefig('analisis_rrhh.png', dpi=150, bbox_inches='tight')
plt.show()
print("Gráfico guardado como analisis_rrhh.png")
```
